<h3>Data-Mining Objective: 1</h3>

```python
#Data-Mining Objective: 1

#Step 1: Assigning our x (independent variables) and y (dependent variable)
X = df.loc[:, df.columns != 'Churn Value']     #X is all columns apart from the "ChurnValue" columns
y = df['Churn Value']

#Step 2: Importing the statsmodels library
import statsmodels.api as sn
import statsmodels.discrete.discrete_model as sm

#Step 3: Adding Constants to X_cons
X_cons = sn.add_constant(X) #adding constant to X_cons (REMEMBER: This was 0 by default)
X_cons.head() #checking our dataframe of x variables
```

```python
#Step 4: Actually running the Logistic Regression Model
# We have already imported the model in Step 2 from the library as "sm"

logit = sm.Logit(y, X_cons).fit() #building our model
```

```python
#Step 5: Get statistical summary
logit.summary()
```

```python
#Step 5: Calculating the odds ratio
    #We know that the odds ratio is Exp(B) where B is the coefficient value

import numpy as np   #importing numpy for exp use

model_odds = pd.DataFrame(np.exp(logit.params), columns = ['Exp(B)'])

print(model_odds)
```

<h3>Data-Mining Objective: 2</h3>

```python
#Step 1: importing essential libraries
from sklearn.model_selection import train_test_split       
from sklearn.linear_model import LogisticRegression
```

```python
#Step 2: train and test split (80:20)

#2A: Generate the training set. Set random_state to be able to replicate results
train = df.sample(frac = 0.8, random_state = 1)

#2B: Select anything not in the training set and put it in the testing set
test = df.loc[~df.index.isin(train.index)]

#2C: Print the shapes of both sets
print(train.shape)
print(test.shape)
```

```python
#Step 3: 

#3A: Creating our classification object:
clf_LR = LogisticRegression() #cls_lrs stands for classification_logisticRegression

#3B: Getting training and testing columns:
x_train = train.loc[:, train.columns != 'Churn Value']     #X is all train dataset columns apart from the "ChurnValue" columns
y_train = train['Churn Value']                             #Y is ChurnValue
x_test  = test.loc[:, train.columns != 'Churn Value']      #X is all test dataset columns apart from the "ChurnValue" columns
y_test  = test['Churn Value']                              #Y is ChurnValue

#3C: Fitting the model to train:
clf_LR.fit(x_train, y_train)
```

```python
#Step 4: Testing the model

#4A: Checking our model on testing dataset
y_test_prediction = clf_LR.predict(x_test)

#4B: Import
from sklearn.metrics import accuracy_score, confusion_matrix

#4C: Confusion_Matrix Testing Dataset
print("Confusion_Matrix")
print(confusion_matrix(y_test, y_test_prediction))
print()
print()

#4D: Accuracy_Score Testing Dataset
print("Accuracy_Score")
print(accuracy_score(y_test, y_test_prediction))
```

```python
#Pattern Search - We want to know the confusion matrix and accuracy score for our training dataset
y_train_prediciton = clf_LR.predict(x_train)

print("Confusion_Matrix")
print(confusion_matrix(y_train, y_train_prediciton))
print()
print()

print("Accuracy_Score")
print (accuracy_score(y_train, y_train_prediciton))
```
